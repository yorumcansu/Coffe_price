```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,5))
plt.plot(kahve.Date,kahve.Close)
plt.title("Coffee Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
kahve1y=kahve.groupby(pd.Grouper(key="Date",freq="1Y")).mean()
kahve1ymax=kahve.groupby(pd.Grouper(key="Date",freq="1Y")).max()
kahve1ymin=kahve.groupby(pd.Grouper(key="Date",freq="1Y")).min()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,5))
plt.bar(kahve1y.index.astype('string'),kahve1ymax.Close,color='green')
plt.bar(kahve1y.index.astype('string'),kahve1y.Close,color='red')
plt.plot(kahve1y.index.astype('string'),kahve1ymin.Close,color='black')
plt.legend(["Min","Ort","Max"])
plt.xlabel("Date")
plt.ylabel('Price - $')
plt.xticks(rotation=61)
plt.title("Coffe Price Per Year")
plt.show()
```
